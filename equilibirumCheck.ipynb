```python
import xarray as xr
import matplotlib.pyplot as plt
from netCDF4 import Dataset
#dataNoAerosol=xr.open_dataset('/scratch/local1/convergenc.nc')
#dataAerosol=xr.open_dataset('/scratch/local1/convergencAerosol.nc')

dataNoAerosol=Dataset('/scratch/local1/m300733/convergence4.nc')
dataAerosol=Dataset('/scratch/local1/m300733/convergenceAerosol4.nc')
```

```python
dataNoAerosol['atmosphere']
```

```python
#dataNoAerosol['radiation']['sw_htngrt']

fig,ax=plt.subplots(3,1)
#plt.suptitle('run without aerosol',fontsize=16)
ax[0].plot(dataAerosol['time'],dataAerosol['radiation']['toa'],label='volcanic aerosol')
ax[0].plot(dataNoAerosol['time'],dataNoAerosol['radiation']['toa'],label='no aerosol')

ax[1].plot(dataAerosol['radiation']['lw_htngrt'][5,:],dataAerosol['plev'],label='volcanic aerosol')
ax[1].plot(dataNoAerosol['radiation']['lw_htngrt'][5,:],dataNoAerosol['plev'],label='no aerosol')
ax[2].plot(dataAerosol['radiation']['sw_htngrt'][5,:],dataAerosol['plev'],label='volcanic aerosol')
ax[2].plot(dataNoAerosol['radiation']['sw_htngrt'][5,:],dataNoAerosol['plev'],label='no aerosol')

ax[0].set_title('radiation budget at top of the atmosphere')
ax[0].set_xlabel('timesteps')
ax[0].set_ylabel('radiation budget [W/m2]')
 
 
#plt.gca().invert_yaxis()
ax[1].set_yscale('log')
ax[1].invert_yaxis()
ax[1].set_title('lw heating rates')
ax[1].set_ylabel('pressure [Pa]')
ax[1].set_xlabel('heating rates [K / day]')

#ax.set_ylim([pressureMin,max(atmosphere['plev'])])
ax[2].set_yscale('log')
ax[2].invert_yaxis()
ax[2].set_title('sw heating rates')
ax[2].set_xlabel('heating rates [K / day]')
ax[2].set_ylabel('pressure [Pa]')

plt.subplots_adjust(bottom=0, left=.01, right=1.99, top=1.90, hspace=.45)
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

```python
fig,ax=plt.subplots(3,1)
#fig.suptitle('run with aerosol')
ax[0].plot(dataAerosol['time'],dataAerosol['radiation']['toa'])
ax[0].set_title('radiation budget at top of the atmosphere')
ax[0].set_xlabel('timesteps')
ax[0].set_ylabel('radiation budget [W/m2]')
ax[1].plot(dataAerosol['radiation']['lw_htngrt'][5,:],dataAerosol['plev'])
 
plt.gca().invert_yaxis()
ax[1].set_yscale('log')
ax[1].invert_yaxis()
ax[1].set_ylabel('pressure [hPa]')
ax[1].set_xlabel('heating rates [K / day]')
ax[1].set_title('lw heating rates')
ax[2].plot(dataAerosol['radiation']['sw_htngrt'][5,:],dataAerosol['plev'])
#ax.set_ylim([pressureMin,max(atmosphere['plev'])])
ax[2].set_yscale('log')
#ax[2].invert_yaxis()
ax[2].set_title('sw heating rates')
ax[2].set_xlabel('heating rates [K / day]')
ax[2].set_ylabel('pressure [Pa]')
plt.subplots_adjust(bottom=0, left=.01, right=1.99, top=1.90, hspace=.45)
plt.show()
```
